```python
from  sklearn.ensemble import RandomForestRegressor
from sklearn import svm, datasets
import pandas as pd 
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler
```

```python
df_features= pd.read_csv("top300_everyting_encoded_and_not_scaled.csv")
df_streams=df_features['Streams']
df_features.drop(['Streams'],axis=1,inplace=True)

df_features
```

```python
import pandas as pd
from sklearn import preprocessing
bin=[3000000,5977762,18671566,50176627,3000000000]
target= pd.cut(df_streams,bin,labels=["Not Popular","Middle","Popular", "Very Popular"])
le = preprocessing.LabelEncoder()
le.fit(target)
target=le.transform(target)

X_train, X_test, y_train, y_test = train_test_split(df_features, target, test_size=0.2)
```

```python
rfr = RandomForestRegressor()
rfr.fit(X_train , y_train) 
rfr.score(X_test , y_test)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron
# Train the scaler, which standarizes all the features to have mean=0 and unit variance
sc = StandardScaler()
sc.fit(X_train)

# Apply the scaler to the X training data
X_train_std = sc.transform(X_train)
# Apply the SAME scaler to the X test data
X_test_std = sc.transform(X_test)

# Create a perceptron object with the parameters: 40 iterations (epochs) over the data, and a learning rate of 0.1
ppn = Perceptron(max_iter=40, eta0=0.1, random_state=0)

# Train the perceptron
ppn.fit(X_train_std, y_train)

# Apply the trained perceptron on the X data to make predicts for the y test data
y_pred = ppn.predict(X_test_std)

print('Accuracy: %.2f' % accuracy_score(y_test, y_pred))
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import MinMaxScaler

X_train = MinMaxScaler().fit_transform(X_train)
mlp = MLPClassifier(verbose=0, random_state=0, max_iter=300, nesterovs_momentum=False,
                    solver='sgd', learning_rate='invscaling', momentum=0.9, learning_rate_init=0.2)

mlp.fit(X_train,y_train)
print("Training set score: %f" % mlp.score(X_train, y_train))
print("Training set loss: %f" % mlp.loss_)
```
